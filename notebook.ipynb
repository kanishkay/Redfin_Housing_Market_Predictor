```python
def distributionCalculator(numApples, numBags):
    applesInBag = numApples / numBags
    return applesInBag

numApples = int(input("Enter the number of apples: "))
numBags = int(input("Enter the number of bags: "))
extraApples = int(input("Enter the number of extra apples: "))
print("No. of apples: ", numApples)
print("No. of bags: ", numBags)
print("Extra apples: ", extraApples)
numApples += extraApples
print("No. of apples received: ", distributionCalculator(numApples, numBags))
print("No. of apples in each bag: " + str(distributionCalculator(numApples, numBags)))
```
